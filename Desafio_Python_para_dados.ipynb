```python
import pandas as pd

df = pd.read_csv('saude_do_sono_estilo_vida.csv')

df.head()
```

1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.

2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]

```python
def estatistica(coluna):
  return coluna.mean().round(1), coluna.median(), coluna.mode()

df.groupby('Profissão')['Duração do sono'].agg(estatistica)
```

3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?

```python
#Filtrando quantidade de pessoas engenheiras de software na coluna profissão
p_eng_software = df['Profissão'] == 'Eng. de Software'
total_p_eng_software = len(df[p_eng_software])

# Filtrando quantidade de pessoas obesas engenheiras de software
filtro_p_obesas_eng_software = (df['Categoria IMC'] == 'Obesidade') & (df['Profissão'] == 'Eng. de Software')
p_obesas_eng_software = len(df[filtro_p_obesas_eng_software])

# Calculando a porcentagem de pessoas obesas engenheiras de software
perc_p_obesas_eng_software = (p_obesas_eng_software / total_p_eng_software) * 100

# Arredondando o valor para duas casas decimais
perc_p_obesas_eng_software = round(perc_p_obesas_eng_software, 2)

# Imprimindo o resultado
print('O percentual de obesos que atuam com engenharia de software é de', perc_p_obesas_eng_software, '%')
```

4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)

```python
# Filtragem das profissões Advogado(a) e Representante de Vendas
filtro_profissao = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]

# Realizando a média das duas profissões e apresentando em 2 casas decimais em ordem descrecente
media_sono = filtro_profissao.groupby('Profissão')['Duração do sono'].mean().sort_values(ascending=True).round(2)

# Imprimindo o resultado
media_sono
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

```python
# Removendo o "/" e dividindo os números da pressão sanguínea
df['Pressão sanguínea'] = df['Pressão sanguínea'].str.replace('/', '')

# Transformando os números em inteiros para realizar o cálculo de média
df['Pressão sanguínea'] = df['Pressão sanguínea'].astype(int)

# Filtrando os dados para obter apenas homens e mulheres
filtro_homens = df[df['Gênero'] == 'Homem']
filtro_mulheres = df[df['Gênero'] == 'Mulher']

# Calculando a média da pressão sanguínea para homens e mulheres
media_pressao_homens = filtro_homens['Pressão sanguínea'].mean()
media_pressao_mulheres = filtro_mulheres['Pressão sanguínea'].mean()

# Comparando e imprimindo a resposta para a pergunta
if media_pressao_homens > media_pressao_mulheres:
    print("Homens têm uma pressão sanguínea média maior.")
elif media_pressao_homens < media_pressao_mulheres:
    print("Mulheres têm uma pressão sanguínea média maior.")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea média.")
```

9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Filtrar frequências cardíacas
acima_70 = df[df['Frequência cardíaca'] > 70]
menor_igual_70 = df[df['Frequência cardíaca'] <= 70]

# Filtrando os passos de acordo com a referência da frequencia cardíaca e tirando a média
media_passos_acima_70 = acima_70['Passos diários'].mean()
media_passos_menor_igual_70 = menor_igual_70['Passos diários'].mean()

# Comparando e imprimindo a resposta para a pergunta
if media_passos_acima_70 > media_passos_menor_igual_70:
    print("Pessoas com frequências cardíacas acima de 70 dão mais passos.")
elif media_passos_acima_70 < media_passos_menor_igual_70:
    print("Pessoas com frequências cardíacas igual ou abaixo de 70 dão mais passos.")
else:
    print("Não há diferença significativa no número de passos entre os dois grupos.")
```
